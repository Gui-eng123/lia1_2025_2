```python

# ---------- 0) Instalar Kaggle CLI ----------
!pip install -q tensorflow keras seaborn scikit-learn kaggle
!pip -q install kaggle

import os, zipfile, shutil, pathlib, tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers # Import the layers module

# ---------- 1) Autenticação Kaggle ----------
# Faça upload do seu kaggle.json para /content/ (Kaggle > Account > Create API Token)
import os, json


kaggle_json = {
  "username": "guilhermecustdio12",
  "key": "6c9d7914fa28c3d2e107c12532c23da"
}


os.makedirs("/root/.kaggle", exist_ok=True)
with open("/root/.kaggle/kaggle.json", "w") as f:
    json.dump(kaggle_json, f)
os.chmod("/root/.kaggle/kaggle.json", 0o600)


# ---------- 2) Escolher dataset e baixar ----------
DATASET = "gpiosenka/cards-image-datasetclassification"   # altere se quiser outro
DOWNLOAD_DIR = "/content"
!kaggle datasets download -d {DATASET} -p {DOWNLOAD_DIR} -q

# ---------- 3) Extrair para pasta organizada ----------
ZIP_PATH = os.path.join(DOWNLOAD_DIR, os.path.basename(DATASET) + ".zip")
EXTRACT_DIR = "/content/kaggle_data"
os.makedirs(EXTRACT_DIR, exist_ok=True)
with zipfile.ZipFile(ZIP_PATH, "r") as z:
    z.extractall(EXTRACT_DIR)

TRAIN_DIR = pathlib.Path(EXTRACT_DIR) / "train"
TEST_DIR  = pathlib.Path(EXTRACT_DIR) / "test"
VALID_DIR = pathlib.Path(EXTRACT_DIR) / "valid"  # se existir, usamos; senão, criamos split

print("Dataset extraído em:", EXTRACT_DIR)
print("Train dir:", TRAIN_DIR)
print("Test  dir:", TEST_DIR)
print("Valid dir:", VALID_DIR if VALID_DIR.exists() else "não existe (usaremos split do train)")

# ---------- 4) Parâmetros do loader ----------
IMG_SIZE   = (224, 224)
BATCH_SIZE = 16
SEED       = 1337
AUTOTUNE   = tf.data.AUTOTUNE

def make_dir_ds(path, shuffle, validation_split=None, subset=None):
    return keras.preprocessing.image_dataset_from_directory(
        path,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        label_mode="int",
        shuffle=shuffle,
        seed=SEED,
        validation_split=validation_split,
        subset=subset
    )

# ---------- 5) Criar train/val/test ----------
if VALID_DIR.exists():
    train_ds = make_dir_ds(TRAIN_DIR, shuffle=True)
    val_ds   = make_dir_ds(VALID_DIR, shuffle=False)
else:
    train_ds = make_dir_ds(TRAIN_DIR, shuffle=True,
                            validation_split=0.15, subset="training")
    val_ds   = make_dir_ds(TRAIN_DIR, shuffle=False,
                            validation_split=0.15, subset="validation")

test_ds = make_dir_ds(TEST_DIR, shuffle=False)

# nomes das classes
nomes_classes = train_ds.class_names
print("Classes:", len(nomes_classes), nomes_classes[:10], "...")

# ---------- 6) Otimizações de I/O (cache em DISCO + prefetch) ----------
CACHE_DIR = "/content/cache_tfdata"
os.makedirs(CACHE_DIR, exist_ok=True)

def optimize(ds, name):
    return ds.cache(os.path.join(CACHE_DIR, f"{name}.cache")).prefetch(AUTOTUNE)

train_ds = optimize(train_ds, "train")
val_ds   = optimize(val_ds,   "val")
test_ds  = optimize(test_ds,  "test")

print("✅ Loader pronto. Use: train_ds, val_ds, test_ds e nomes_classes")


# ---------- 7) Criar/compilar o modelo (CNN simples 224x224x3) ----------

modelo = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(nomes_classes), activation='softmax')
])

modelo.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

modelo.summary()

# Visualização classes
# =============================
import matplotlib.pyplot as plt

def visualiza_tfdata(dataset, nomes, n=25):
    plt.figure(figsize=(10,10))
    for images, labels in dataset.take(1):   # pega 1 batch do dataset
        for i in range(min(n, images.shape[0])):
            plt.subplot(5,5,i+1)
            plt.xticks([]); plt.yticks([]); plt.grid(False)
            plt.imshow(images[i].numpy().astype("uint8"))
            plt.xlabel(nomes[int(labels[i])])
    plt.tight_layout()
    plt.show()

visualiza_tfdata(train_ds, nomes_classes)



# ============================================================
# TREINO + AVALIAÇÃO + MATRIZ DE CONFUSÃO
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau


callbacks = [
    EarlyStopping(monitor="val_accuracy", patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_accuracy", factor=0.5, patience=2, verbose=1),
    ModelCheckpoint("/content/melhor_modelo.keras", monitor="val_accuracy",
                    save_best_only=True, verbose=1)
]


# ---------- 2) Treinamento ----------
EPOCHS = 8
hist = modelo.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

# ---------- 3) Curvas de treino ----------
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist.history["accuracy"], label="treino")
plt.plot(hist.history["val_accuracy"], label="val")
plt.title("Acurácia"); plt.xlabel("Épocas"); plt.ylabel("acc"); plt.legend()

plt.subplot(1,2,2)
plt.plot(hist.history["loss"], label="treino")
plt.plot(hist.history["val_loss"], label="val")
plt.title("Loss"); plt.xlabel("Épocas"); plt.ylabel("loss"); plt.legend()
plt.tight_layout(); plt.show()

# ---------- 4) Avaliação em TESTE ----------
test_metrics = modelo.evaluate(test_ds, verbose=0)
# Mantém compatível com modelos que retornam (loss, acc) ou mais métricas
metric_names = [m.name if hasattr(m, "name") else str(m) for m in modelo.metrics]
metric_report = dict(zip(["loss"] + metric_names, test_metrics))
print("Métricas (teste):", metric_report)

# ---------- 5) Matriz de confusão + relatório ----------
y_true, y_pred = [], []
for batch_imgs, batch_labels in test_ds:
    probs = modelo.predict(batch_imgs, verbose=0)
    y_pred.extend(np.argmax(probs, axis=1))
    y_true.extend(batch_labels.numpy())


y_true = np.array(y_true, dtype=int)
y_pred = np.array(y_pred, dtype=int)

cm = confusion_matrix(y_true, y_pred)
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

THRESH = 0.10
mask_annot = cm_norm >= THRESH
annot = np.where(mask_annot, (cm_norm*100).round(1).astype(str), "")

plt.figure(figsize=(21,21))  # aumenta área da figura
sns.heatmap(
    cm_norm, cmap="Blues",
    xticklabels=nomes_classes, yticklabels=nomes_classes,
    annot=annot, fmt="", linewidths=.2, linecolor="#eee",
    cbar_kws={"label": "% por classe real"}
)

plt.xlabel("Previsto", fontsize=16)
plt.ylabel("Real", fontsize=16)
plt.title("Matriz de Confusão normalizada", fontsize=17)

plt.xticks(rotation=90, fontsize=10)
plt.yticks(rotation=0, fontsize=10)

plt.tight_layout()
plt.show()

# ---------- 6) Salvar artefatos ----------
modelo.save("/content/modelo_tfdata.keras")
import json
with open("/content/classes.json","w",encoding="utf-8") as f:
    json.dump(nomes_classes, f, ensure_ascii=False)

print("Treinamento concluído")
```

```python

# 2ª CÉLULA — Upload + Inferência
import io
from PIL import Image
from google.colab import files

def _preprocess(img_pil, img_size):
    """PIL -> tensor [1,H,W,3] float32 normalizado 0..1 (igual ao treino)."""
    img = img_pil.convert("RGB").resize(img_size, Image.Resampling.BILINEAR)
    x = np.array(img, dtype=np.float32)
    return np.expand_dims(x, axis=0)

def _predict(img_pil, model, class_names, img_size, topk=3):
    x = _preprocess(img_pil, img_size)
    probs = model.predict(x, verbose=0)[0]           # shape: [num_classes]
    idx = np.argsort(probs)[::-1][:topk]
    return [(class_names[i], float(probs[i])) for i in idx]  # [(label, conf), ...]

def _show(img_pil, top, title="Predição"):
    plt.figure(figsize=(5,5))
    plt.imshow(img_pil); plt.axis("off"); plt.title(title)
    plt.show()
    print("Top previsões:")
    for lbl, p in top:
        print(f"  - {lbl:25s} {p*100:6.2f}%")

def classificar_imagens(modelo, nomes_classes, IMG_SIZE=(224,224), topk=3, low_conf=0.50):
    uploaded = files.upload()  # selecione 1..N arquivos
    for fname, content in uploaded.items():
        print("\n==============================")
        print(f"Arquivo: {fname}")
        img = Image.open(io.BytesIO(content)).convert("RGB")
        top = _predict(img, modelo, nomes_classes, IMG_SIZE, topk=topk)
        lbl1, conf1 = top[0]
        _show(img, top, title=f"{lbl1} ({conf1*100:.2f}%)")
        if conf1 < low_conf:
            print(f"⚠️ Confiança baixa ({conf1*100:.2f}%). Tente foto mais centralizada/nítida.")

# --- Executar:
# classificar_imagens(modelo, nomes_classes, IMG_SIZE=IMG_SIZE, topk=3, low_conf=0.50)
classificar_imagens(modelo, nomes_classes, IMG_SIZE=IMG_SIZE, topk=3, low_conf=0.50)
```
